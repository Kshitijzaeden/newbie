```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(9,6))

plt.title("Gas price over time in USD",fontdict={'fontname':'Comic Sans Ms','fontsize':20})

for country in gas:
  if country!="Year":
    plt.plot(gas.Year,gas[country],marker='o')



plt.xticks(gas.Year[::3])
plt.xlabel("Years",fontdict={'fontname':'Comic Sans Ms','fontsize':20})
plt.ylabel("US Dollars ",fontdict={'fontname':'Comic Sans Ms','fontsize':20})

plt.legend()
plt.show()
```
