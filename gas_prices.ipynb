```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv('gas_prices.csv')
country=pd.read_csv('gas_prices.csv')

plt.figure(figsize=(9,6))

plt.plot(gas.Year,gas.USA,label='USA')
plt.plot(gas.Year,gas.Australia,label='Australia')
plt.plot(gas.Year,gas.Canada,label='Canada')
plt.plot(gas.Year,gas.France,label='France')
plt.plot(gas.Year,gas.Germany,label='Germany')
plt.plot(gas.Year,gas.Italy,label='Italy')
plt.plot(gas.Year,gas.Japan,label='Japan')
plt.plot(gas.Year,gas.Mexico,label='Mexico')
plt.plot(gas.Year,gas['South Korea'],label='South Korea')
plt.plot(gas.Year,gas.UK,label='UK')

plt.title("Gas price over time in USD",fontdict={'fontname':'Aerial black','fontsize':20})

for country in gas:
  if country!="Year":
    plt.plot(gas.Year,gas[country],marker='o')

plt.xticks(gas.Year[::1])
plt.xlabel("Years",fontsize=20)
plt.ylabel("US Dollars ",fontsize=20)

plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
